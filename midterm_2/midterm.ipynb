##### **Otros valores de *N***

3.8 - Repita los incisos del *3.1* al *3.7* con las siguientes cantidades de *n*:
* 10K nodos
* 25k nodos


```python
def run_bst_avl_analysis(n_size, insert_values, search_values):
    print(f"\nAnálisis para {n_size} nodos: ")

    bst = BinarySearchTree()
    avl = AVLTree()
    for i in range(1, n_size + 1):
        bst.insert(i)
        avl.insert(i)

    # Insert
    bst_insert_times = []
    avl_insert_times = []
    for val in insert_values:
        bst_insert_times.append(time_function(bst.insert, val))
        avl_insert_times.append(time_function(avl.insert, val))

    # Search  
    bst_search_times = []
    avl_search_times = []
    for val in search_values:
        bst_search_times.append(time_function(bst.search, val))
        avl_search_times.append(time_function(avl.search, val))

    # Promedios
    bst_insert_avg = sum(bst_insert_times) / len(bst_insert_times)
    avl_insert_avg = sum(avl_insert_times) / len(avl_insert_times)
    bst_search_avg = sum(bst_search_times) / len(bst_search_times)
    avl_search_avg = sum(avl_search_times) / len(avl_search_times)

    # Resultados
    print(f"BST Insert Avg: {bst_insert_avg:.10f} secs")
    print(f"AVL Insert Avg: {avl_insert_avg:.10f} secs")
    print(f"BST Search Avg: {bst_search_avg:.10f} secs")
    print(f"AVL Search Avg: {avl_search_avg:.10f} secs")

    return {
        "bst_insert": bst_insert_avg,
        "avl_insert": avl_insert_avg,
        "bst_search": bst_search_avg,
        "avl_search": avl_search_avg
    }
```

```python

Insert_BIG_O_Values = list(range(50000, 50025))
Search_BIG_O_Values = list(range(100000, 100025))

# Ejecutar para 10K y 25K
results_10k = run_bst_avl_analysis(10000, Insert_BIG_O_Values, Search_BIG_O_Values)
results_25k = run_bst_avl_analysis(25000, Insert_BIG_O_Values, Search_BIG_O_Values)
```
